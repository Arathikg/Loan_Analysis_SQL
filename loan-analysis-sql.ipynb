The average, minimum and maximum income of applicants who were applied for a loan

```python
%%sql 
SELECT 
    Loan_Status,
    ROUND(AVG(ApplicantIncome), 2) AS avg_applicant_income,
    MIN(ApplicantIncome) AS min_applicant_income,
    MAX(ApplicantIncome) AS max_applicant_income
FROM loan_data
GROUP BY 1;
```

what is the percentage of such loan amount getting approved and disapproved?

```python


       %%sql
SELECT property_area, loan_status,sum(loanamount), COUNT(*) AS count
FROM loan_data
WHERE loanamount > (
    SELECT AVG(loanamount)
    FROM loan_data
)
GROUP BY property_area, loan_status
ORDER BY loanamount DESC, property_area DESC;




```

```python
%%sql
SELECT property_area, loan_status, loanamount, COUNT(*) AS count
FROM loan_data
WHERE loanamount > (
    SELECT AVG(loanamount)
    FROM loan_data
)
GROUP BY property_area, loan_status, loanamount
ORDER BY loanamount DESC, property_area DESC;
```

```python
%%sql
SELECT 
    Dependents,
    COUNT(*)  AS total_applicants,
    SUM(CASE WHEN Loan_Status = 'Y' THEN 1 ELSE 0 END) AS approved_applicants,
    ROUND(AVG(CASE WHEN Loan_Status = 'Y' THEN 1 ELSE 0 END) * 100,2) loan_approval_rate
FROM loan_data
GROUP BY 1;
```

 What is the count of loans for each property area, and how does this compare to the total count of loans?

```python
%%sql
SELECT
    Property_Area,
    COUNT(*) AS frequency,
    total_loans
FROM loan_data
JOIN (
    SELECT COUNT(*) AS total_loans
    FROM loan_data
) AS total_counts
GROUP BY Property_Area, total_loans;
```
